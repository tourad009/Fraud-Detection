```python
import category_encoders as ce

cols_to_encode = ['ProviderId', 'ProductId', 'ChannelId', 'ProductCategory']

# Séparer X (features) et y (target)
X= df_train.drop(columns=['FraudResult'])
y = df_train['FraudResult']

# Créer l'encodeur
encoder = ce.OrdinalEncoder(cols=cols_to_encode)
X_encoded = encoder.fit_transform(X)

# Pour df_test
df_test_encoded = encoder.transform(df_test)
```

```python
X_encoded.head()
```

```python
df_test_encoded
```

```python
# Assurer que la colonne est bien en datetime
df_train = pd.concat([X_encoded, y], axis=1)

df_train['TransactionStartTime'] = pd.to_datetime(df_train['TransactionStartTime'])
df_test['TransactionStartTime'] = pd.to_datetime(df_test['TransactionStartTime'])

# Extraire les caractéristiques temporelles
df_train['Month'] = df_train['TransactionStartTime'].dt.month
df_train['Hour'] = df_train['TransactionStartTime'].dt.hour
df_train['Weekday'] = df_train['TransactionStartTime'].dt.weekday  # 0 = Lundi, 6 = Dimanche

df_test['Month'] = df_test['TransactionStartTime'].dt.month
df_test['Hour'] = df_test['TransactionStartTime'].dt.hour
df_test['Weekday'] = df_test['TransactionStartTime'].dt.weekday  # 0 = Lundi, 6 = Dimanche

# Supprimer la colonne TransactionStartTime si plus nécessaire
df_train.drop(columns=['TransactionStartTime'], inplace=True)
df_test.drop(columns=['TransactionStartTime'], inplace=True)
```

```python
df_train.head()
```

```python
X_encoded = df_train.drop('FraudResult', axis=1)
X_encoded.head()
```

```python
df_test
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# 1. Séparation Train / Validation
X_train, X_val, y_train, y_val = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# 3. Modèle de base
rf_model = RandomForestClassifier(
    max_depth=15,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
# 4. Prédiction et évaluation sur Validation (non resamplé)
y_pred = rf_model.predict(X_val)

print(classification_report(y_val, y_pred, target_names=['Not Fraud', 'Fraud']))
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report

# 1. Modèle CatBoost ultra simple + boost léger
cat_model = CatBoostClassifier(
    random_state=42,
    iterations=100,
    learning_rate=0.415,
    loss_function='Logloss',
    verbose=0,
)

# 2. Entraînement
cat_model.fit(X_train, y_train)

# 3. Prédictions
y_pred = cat_model.predict(X_val)

# 4. Rapport
print("\nClassification Report:\n")
print(classification_report(y_val, y_pred, target_names=['Not Fraud', 'Fraud']))
```

```python
import lightgbm as lgb
from sklearn.metrics import classification_report, f1_score
import numpy as np

# 1. Modèle LightGBM simple mais costaud
lgbm_model = lgb.LGBMClassifier(
    objective='binary',
    boosting_type='gbdt',
    max_depth=10,              # contrôle la complexité
    num_leaves=31,             # standard efficace pour n_samples moyens
    learning_rate=0.05,        # pas trop agressif
    n_estimators=500,          # assez pour bien apprendre
    class_weight='balanced',   # très important ici
    random_state=42,
    n_jobs=-1                  # entraînement rapide
)

# 2. Entraînement
lgbm_model.fit(X_train, y_train)

# Probabilités de classe 1
y_proba = lgbm_model.predict_proba(X_val)[:, 1]

# Chercher le meilleur threshold
best_threshold = 0.5
best_f1 = 0

thresholds = np.arange(0.1, 0.9, 0.01)

for thresh in thresholds:
    y_pred_thresh = (y_proba >= thresh).astype(int)
    f1 = f1_score(y_val, y_pred_thresh)
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = thresh

print(f"✅ Best threshold: {best_threshold:.2f} with F1-score: {best_f1:.4f}")

# Appliquer le meilleur threshold
y_pred_best = (y_proba >= best_threshold).astype(int)

# Nouveau rapport de classification
print("\nClassification Report with Best Threshold:\n")
print(classification_report(y_val, y_pred_best, target_names=['Not Fraud', 'Fraud']))
```

```python
test_data = pd.concat([df_test_encoded, df_test[['Month', 'Hour', 'Weekday']]], axis=1).drop('TransactionStartTime', axis=1)
test_data
```

```python
y_pred_test = lgbm_model.predict(test_data)

# Créer le dataframe de soumission
submission = pd.DataFrame({
    'TransactionId': pd.read_csv('../data/test.csv')['TransactionId'],
    'FraudResult': y_pred_test
})

print("Distribution des prédictions avant soumission :")
print(submission['FraudResult'].value_counts())

# Sauvegarder
submission.to_csv('submission_cta.csv', index=False)
```
